## 1. Conditional Statements
## Conditional statements allow decision-making in programs based on conditions.
Definition

A conditional statement is a control structure that allows the program to execute different blocks of code based on the evaluation of a Boolean expression.

The condition must evaluate to:

True → execute associated block

False → skip or execute alternative block

```python


# Conditional statements are also called:

# Decision-making statements

# Selection structures

# Branching structures

# They rely on:

# Relational operators
# > , < , >= , <= , == , !=

# Logical operators
# and , or , not
```

# Types of Conditional Statements in Python

Python provides four types of conditional structures:

if

if-else

if-elif-else

Nested if

### if Statement
Executes a block only when condition is true.
The if statement executes a block of code only when the condition is true.

marks >= 40 and attendance >= 75

```python
age = 20
if age >= 18:
    print("Eligible to vote")
```

### if-else Statement
Provides two paths depending on condition.

```python
num = 7
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

### if-elif-else Ladder
Used for multiple conditions.

```python
marks = 78
if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Fail")
```

### Nested if Statement
One conditional inside another.

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login successful")
    else:
        print("Wrong password")
else:
    print("Invalid user")
```

## 2. Loops
Loops allow repetition of a block of code.



A loop is a control structure that repeatedly executes a block of code as long as a specified condition remains true.

Loops are used to:

Reduce redundancy

Process collections of data

Perform repetitive computations

Automate tasks

Loops are categorized as:

Type	Description
Definite loop	Number of iterations is known
Indefinite loop	Number of iterations depends on condition

Python provides:

for loop → Definite iteration

while loop → Indefinite iteration

```python
for i in range(1, 6):
    print(i)
```

### while Loop
Used when iterations depend on a condition.

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

## 3. Loop Control Statements

### break Statement
Stops loop immediately.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

### continue Statement
Skips current iteration.

```python
for i in range(6):
    if i == 2:
        continue
    print(i)
```

### pass Statement
Placeholder that does nothing.

```python
for i in range(5):
    pass
print("Loop completed")
```

## 4. Combined Example

```python
customers = [1500, 4200, 800, 6000, 3500]

for amount in customers:
    if amount >= 3000:
        print(amount, "- Premium Customer")
    else:
        print(amount, "- Regular Customer")
```
