# Python Operators and Expressions

This notebook covers all major types of operators in Python with definitions and executable examples.

## 1. Operators and Expressions
**Operators** are special symbols in Python that perform operations on variables and values.

**Expressions** are combinations of variables, operators, and values that produce a result.

## 2. Arithmetic Operators
Arithmetic operators perform mathematical calculations.

| Operator | Description | Example | Output |
|--------|--------------|---------|--------|
| + | Addition | 5 + 3 | 8 |
| - | Subtraction | 10 - 4 | 6 |
| * | Multiplication | 6 * 2 | 12 |
| / | Division | 7 / 2 | 3.5 |
| % | Modulus | 7 % 2 | 1 |
| // | Floor Division | 7 // 2 | 3 |
| ** | Exponentiation | 2 ** 3 | 8 |

```python
a, b = 10, 3
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a % b)
print(a // b)
print(a ** b)
```

## 3. Comparison Operators
Comparison operators compare values and return True or False.

```python
x, y = 10, 20
print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= y)
print(x <= y)
```

## 4. Logical Operators
Logical operators are used to combine multiple conditions: and, or, not.

```python
a, b = True, False
print(a and b)
print(a or b)
print(not a)
```

## 5. Bitwise Operators
Bitwise operators perform operations at the binary level.

```python
x, y = 5, 3
print(x & y)
print(x | y)
print(x ^ y)
print(~x)
print(x << 1)
print(x >> 1)
```

## 6. Assignment Operators
Assignment operators are used to assign and update values.

```python
x = 10
x += 5
print(x)
x *= 2
print(x)
```

## 7. Identity Operators
`is` and `is not` check whether two variables point to the same object.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)
print(a is c)
print(a == c)
```

## 8. Membership Operators
`in` and `not in` test whether a value exists in a sequence.

```python
numbers = [1, 2, 3, 4]
print(2 in numbers)
print(5 not in numbers)
```

## 9. Operator Precedence
Order of execution (highest to lowest):

1. Parentheses ()
2. Exponentiation **
3. Unary +, -, ~
4. *, /, //, %
5. +, -
6. <<, >>
7. &
8. ^
9. |
10. Comparisons
11. not
12. and
13. or
14. Assignment

```python
print(2 + 3 * 4)
print((2 + 3) * 4)
```
