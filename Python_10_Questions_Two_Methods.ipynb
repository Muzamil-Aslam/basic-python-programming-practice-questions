## 1. Sum of two numbers

### Problem:
Write a program to input two numbers and print their sum.

```python
# Method 1: Using simple addition
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum:", a + b)
```

```python
# Method 2: Using a function
def add(x, y):
    return x + y

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("Sum:", add(a, b))
```

**Explanation:**
- Method 1 directly adds two numbers.
- Method 2 uses a function to make the code reusable.

## 2. Check even or odd

### Problem:
Check whether a number is even or odd.

```python
# Method 1: Using modulus operator
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Method 2: Using bitwise operator
num = int(input("Enter a number: "))

if num & 1 == 0:
    print("Even")
else:
    print("Odd")
```

**Explanation:**
- `% 2` checks remainder.
- `& 1` checks the last binary bit (efficient method).

## 3. Largest of three numbers

### Problem:
Find the largest of three numbers.

```python
# Method 1: Using max()
a = int(input())
b = int(input())
c = int(input())

print("Largest:", max(a, b, c))
```

```python
# Method 2: Using if-else
a = int(input())
b = int(input())
c = int(input())

if a > b and a > c:
    print("Largest:", a)
elif b > c:
    print("Largest:", b)
else:
    print("Largest:", c)
```

**Explanation:**
- `max()` is built-in.
- `if-else` shows logical comparison approach.

## 4. Print numbers from 1 to 100

### Problem:
Print numbers from 1 to 100.

```python
# Method 1: Using for loop
for i in range(1, 101):
    print(i)
```

```python
# Method 2: Using while loop
i = 1
while i <= 100:
    print(i)
    i += 1
```

**Explanation:**
- Both loops repeat code.
- `for` is compact, `while` gives manual control.

## 5. Factorial of a number

### Problem:
Find factorial of a number.

```python
# Method 1: Using loop
num = int(input())
fact = 1

for i in range(1, num+1):
    fact *= i

print("Factorial:", fact)
```

```python
# Method 2: Using recursion
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

num = int(input())
print("Factorial:", factorial(num))
```

**Explanation:**
- Loop is iterative.
- Recursion uses function calling itself.

## 6. Reverse a string

### Problem:
Reverse a string.

```python
# Method 1: Using slicing
s = input()
print(s[::-1])
```

```python
# Method 2: Using loop
s = input()
rev = ""

for ch in s:
    rev = ch + rev

print(rev)
```

**Explanation:**
- Slicing is fastest.
- Loop shows step-by-step logic.

## 7. Check palindrome

### Problem:
Check whether a string is palindrome.

```python
# Method 1: Using slicing
s = input()

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# Method 2: Using loop
s = input()
rev = ""

for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not palindrome")
```

**Explanation:**
- Compare original with reversed string.

## 8. Count vowels in a string

### Problem:
Count vowels in a string.

```python
# Method 1: Using loop
s = input()
count = 0

for ch in s.lower():
    if ch in "aeiou":
        count += 1

print(count)
```

```python
# Method 2: Using list comprehension
s = input()

vowels = [ch for ch in s.lower() if ch in "aeiou"]
print(len(vowels))
```

**Explanation:**
- Loop counts one by one.
- List comprehension is compact.

## 9. Sum of elements in a list

### Problem:
Find sum of list elements.

```python
# Method 1: Using sum()
nums = [10, 20, 30, 40]
print(sum(nums))
```

```python
# Method 2: Using loop
nums = [10, 20, 30, 40]
total = 0

for n in nums:
    total += n

print(total)
```

**Explanation:**
- Built-in vs manual addition.

## 10. Remove duplicates from a list

### Problem:
Remove duplicate elements from a list.

```python
# Method 1: Using set
nums = [1, 2, 2, 3, 4, 4]
unique = list(set(nums))
print(unique)
```

```python
# Method 2: Using loop
nums = [1, 2, 2, 3, 4, 4]
unique = []

for n in nums:
    if n not in unique:
        unique.append(n)

print(unique)
```

**Explanation:**
- `set()` is easy but unordered.
- Loop preserves order.
